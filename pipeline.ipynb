```python
import os
import shutil
from pathlib import Path

src_dir = Path("data/table_data/report_table_data/1month")
dst_dir = Path("data/datatales")
dst_dir.mkdir(parents=True, exist_ok=True)

for split in ["train", "validate", "test"]:
    split_dir = src_dir / split
    if not split_dir.exists():
        continue
    for market_dir in split_dir.iterdir():
        if not market_dir.is_dir():
            continue
        
        target_dir = dst_dir / market_dir.name / split
        target_dir.mkdir(parents=True, exist_ok=True)
        
        for csv_file in market_dir.glob("*.csv"):
            shutil.copy(csv_file, target_dir / csv_file.name)
```
